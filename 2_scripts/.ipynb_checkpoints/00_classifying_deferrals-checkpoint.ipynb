```python
import pyreadstat
import pandas as pd
import numpy as np
import datetime as dt
from datetime import timedelta

import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
# a function to compute deferral rates and make a 2X2 table

def make_table(df):
    first=df.query("first_time == 1")
    repeat=df.query("first_time == 0")
    
    #create a dict
    table={}
    table['-']=['Female', 'Male', 'All']

        #first time
    female=(len(first.loc[(first.DONOR_GENDER == 'F') & (first.OUTCOME_TYPE=='low hgb')])/len(first.loc[(first.DONOR_GENDER == 'F')]))
    male=(len(first.loc[(first.DONOR_GENDER == 'M') & (first.OUTCOME_TYPE=='low hgb')])/len(first.loc[(first.DONOR_GENDER == 'M')]))
    alls=(len(first.loc[(first.OUTCOME_TYPE=='low hgb')])/len(first))
    table['First Time']=[female, male, alls]

        #repeat
    female=(len(repeat.loc[(repeat.DONOR_GENDER == 'F') & (repeat.OUTCOME_TYPE=='low hgb')])/len(repeat.loc[(repeat.DONOR_GENDER == 'F')]))
    male=(len(repeat.loc[(repeat.DONOR_GENDER == 'M') & (repeat.OUTCOME_TYPE=='low hgb')])/len(repeat.loc[(repeat.DONOR_GENDER == 'M')]))
    alls=(len(repeat.loc[(repeat.OUTCOME_TYPE=='low hgb')])/len(repeat))
    table['Repeat']=[female, male, alls]

        #all 
    female=(len(df.loc[(df.DONOR_GENDER == 'F') & (df.OUTCOME_TYPE=='low hgb')])/len(df.loc[(df.DONOR_GENDER == 'F')]))
    male=(len(df.loc[(df.DONOR_GENDER == 'M') & (df.OUTCOME_TYPE=='low hgb')])/len(df.loc[(df.DONOR_GENDER == 'M')]))
    alls=(len(df.loc[(df.OUTCOME_TYPE=='low hgb')])/len(df))
    table['All']=[female, male, alls]
    table=pd.DataFrame(table)
    
    return table
    
```

```python
#function to make a crosstable of deferral groups

def get_cross_tab(df1):
    
    #if hgb value is less than threshold and def_group has low_hemoglobin
    df1['DEF_GROUP_LOW_HGB']=df1.apply(lambda x: int((x['OUTCOME_TYPE'] =='low hgb') & ('Low Hemoglobin' in x['DEFERRAL_GROUP_y'])), axis=1)

    #if hgb value is less than threshold and def_group has blood loss
    df1['DEF_GROUP_BL'] = df1.apply(lambda x: int((x['OUTCOME_TYPE'] =='low hgb') & ('Blood Loss' in x['DEFERRAL_GROUP_y'])), axis=1)
    #fill nans
    df1['DEF_GROUP_LOW_HGB'].fillna(0, inplace=True)
    df1['DEF_GROUP_BL'].fillna(0, inplace=True)

    #if hgb value is less than threshold and def_group has both, low hgb and blood loss
    df1['DEF_GROUP_LH_or_BL'] = df1['DEF_GROUP_LOW_HGB']*df1['DEF_GROUP_BL']

    return df1
    
#df1['OUTCOME_TYPE_4']=np.where((df1.OUTCOME_TYPE_3 ==1) and (list>2), 1,0)
```

```python
#load relevant data

don=pd.read_csv("../1_data/private/donation_donor.csv") #donations
defr=pd.read_csv("../1_data/private/deferral.csv", error_bad_lines=False) #deferrals
med_quest=pd.read_csv("../1_data/private/medical_quest.csv", error_bad_lines=False) #hgb value
```

```python
med_questHGB=med_quest.loc[(med_quest['QUESTION_CODE']=='5U')] #df with hgb values
med_questHGB_recheck=med_quest.loc[(med_quest['QUESTION_CODE']=='5V')] #df with hgb recheck values
```

```python
#using only donations table linked to med_quest for hgb values

#add hgb values
df=pd.merge(don, med_questHGB[["DONOR_NUMBER", "DON_DATE_KEY", "DONATION_NUMBER", 'MQ_ANSWER']], how='left', on=["DONOR_NUMBER","DONATION_NUMBER", "DON_DATE_KEY" ]) 
#add hgb recheck values
df=pd.merge(df, med_questHGB_recheck[["DONOR_NUMBER", "DON_DATE_KEY", "DONATION_NUMBER", 'MQ_ANSWER']], how='left', on=["DONOR_NUMBER", "DONATION_NUMBER", "DON_DATE_KEY"])
print(len(df))

#define hgb value-> if recheck if not na replace hgb value with recheck value
df['HGB_value']=np.where((df.MQ_ANSWER_y.notna()),df.MQ_ANSWER_y, df.MQ_ANSWER_x)# if hgb recheck avaible use that as hgb val else use hgb val

#convert MQ_answer to float and replace nans
df['HGB_value']=pd.to_numeric(df['HGB_value'], errors='coerce')
df['HGB_value'].replace(np.nan, 999, regex=True, inplace=True)

df['first_time']=np.where(df.ACTIVE_DONOR_TYPE=='First Time', 1,0) #add first time col

#check if hgb value is below threshold
df.loc[(df['HGB_value']<=13.0) & (df['DONOR_GENDER'] == "M"), 'OUTCOME_TYPE'] = 'low hgb'
df.loc[(df['HGB_value']<=12.5) & (df['DONOR_GENDER'] == "F"), 'OUTCOME_TYPE'] = 'low hgb'

#make table
table=make_table(df)
print(table)
df['OUTCOME_TYPE'].value_counts()
```

The above table shows the Deferral Rate computed using only donations table to categorise low hgb using hgb value


```python
#using a merged data - donations with deferrals (left join)

df1=pd.merge(don, defr, how='left', on=["DONOR_KEY", "DON_DATE_KEY", 'DONATION_NUMBER'])
print(len(df1))

df1=pd.merge(df1, med_questHGB[["DONOR_NUMBER", "DON_DATE_KEY", "DONATION_NUMBER", 'MQ_ANSWER']], how='left', on=["DONOR_NUMBER","DONATION_NUMBER", "DON_DATE_KEY" ])

df1=pd.merge(df1, med_questHGB_recheck[["DONOR_NUMBER", "DON_DATE_KEY", "DONATION_NUMBER", 'MQ_ANSWER']], how='left', on=["DONOR_NUMBER", "DONATION_NUMBER", "DON_DATE_KEY"])

df1['HGB_value']=np.where((df1.MQ_ANSWER_y.notna()),df1.MQ_ANSWER_y, df1.MQ_ANSWER_x)# if hgb recheck avaible use that as hgb val else use hgb val

#convert MQ_answer to float
df1['HGB_value']=pd.to_numeric(df1['HGB_value'], errors='coerce')
#replace all nan hgb values with 999
df1['HGB_value'].replace(np.nan, 999, regex=True, inplace=True)

df1['first_time']=np.where(df1.ACTIVE_DONOR_TYPE=='First Time', 1,0)

df1.loc[(df1['HGB_value']<=13.0) & (df1['DONOR_GENDER'] == "M"), 'OUTCOME_TYPE'] = 'low hgb'
df1.loc[(df1['HGB_value']<=12.5) & (df1['DONOR_GENDER'] == "F"), 'OUTCOME_TYPE'] = 'low hgb'

table=make_table(df1)
print(table)
df1['OUTCOME_TYPE'].value_counts()
```


The above table shows the Deferral Rate computed using donations table merged (left join) with deferrals table on the following:
"DONOR_KEY", "DON_DATE_KEY", 'DONATION_NUMBER'

Then I used the hgb value to categorise low hgb outcomes


```python
#group deferral by don_key and don_date key

def_unique=defr.groupby(['DONOR_KEY', "DON_DATE_KEY"])["DEFERRAL_GROUP"].apply(list)
print(len(def_unique))
```
