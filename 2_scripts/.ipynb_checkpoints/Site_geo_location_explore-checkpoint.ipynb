```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim, GoogleV3
from geopy.point import Point
```

```python
df=pd.read_csv("../1_data/private/Site_Geo_location.csv")
df.head() 
```

```python
percent_missing = df.isna().sum() * 100 / len(df)
percent_missing
```

```python
#find address using longitutde and latitude

# Initialize Nominatim API
geolocator=Nominatim(user_agent='geoapiExercises')

#get location using function geocode
df['address']= np.where((df['lat'].notna() & df['long'].notna()), 
                        geolocator.reverse(Point(df['lat'].all(), df['long'].all())), '')

df.head()
```
