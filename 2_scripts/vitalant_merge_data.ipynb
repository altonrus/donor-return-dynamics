```python
import pyreadstat
import pandas as pd
import numpy as np
import datetime as dt
from datetime import timedelta

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df, meta = pyreadstat.read_sas7bdat("../1_data/private/vitalant/donation_donor.sas7bdat")
don=pd.read_csv("../1_data/private/donation_donor.csv", error_bad_lines=False)
defr=pd.read_csv("../1_data/private/deferral.csv", error_bad_lines=False)
collect=pd.read_csv("../1_data/private/collec_sites.csv", error_bad_lines=False)

med_quest=pd.read_csv("../1_data/private/medical_quest.csv", error_bad_lines=False)
```

```python
med_quest=med_quest.loc[(med_quest['QUESTION_CODE']=='5U')|(med_quest['QUESTION_CODE']=='5V')]
```

```python
#med_quest.head(25)
```

```python
# merge don with collect

#merge donations and deferrals
df=pd.merge(don, collect, how='left', on=["COLLECTION_SITE_KEY"])

df.head()
```

```python
#merge df with deferrals
df1=pd.merge(df, defr, how='left', on=["DONOR_KEY", "DON_DATE_KEY", "DONATION_NUMBER"])

df1.head()
```

```python
#merge hemoglobin level from med_quest
df1=pd.merge(df1, med_quest, how='left', on=["DONOR_NUMBER", "DON_DATE_KEY", "DONATION_NUMBER"])

df1.head()
```
